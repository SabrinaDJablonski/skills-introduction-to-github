##Funciones en Python

Una función en Python (y en cualquier otro lenguaje de programación) es un bloque de líneas de código o un conjunto de instrucciones cuya finalidad es realizar una tarea específica.  Nos ayuda a que el código sea más fácil de leer y entender.

Sintáxis de una función:

image.png

Para definir una función nueva en Python utiliza la palabra reservada **def**, seguida del nombre que le querés dar a la función. A continuación, se agregan los paréntesis con los parámetros de entrada de la función seguidos de dos puntos **(:)**. Después, normalmente en otra línea de código, se agrega el bloque de instrucciones que querés que se ejecuten cada vez que se invoca a la función.

Dentro de este bloque de instrucciones (casi siempre al final y de forma opcional) se coloca una sentencia de retorno que no es más que el valor que deseas que devuelva la función: para esto usa la palabra reservada **return**.

Una función en Python siempre devolverá un valor; es decir, si no se agrega una sentencia de retorno (return), se devolverá el valor predeterminado **None**.

```python
def suma(a, b):

    return a + b

suma(8,5)
```

```python
def sum(num1: int, num2: int) -> int:  #especificar el tipo de datos (opcional)
    """Add two numbers""" ##docstring: se usa para describir la funcionalidad
    num3 = num1 + num2

    return num3


num1, num2 = 5, 15
ans = sum(num1, num2)
print(f"{num1} más {num2} da {ans}.")
```

```python
def evenOdd(x):
    if (x % 2 == 0):
        print("even")
    else:
        print("odd")


# Driver code to call the function
evenOdd(2)
evenOdd(3)
```

```python
def square_value(num):
    """This function returns the square
    value of the entered number"""
    return num**2


print(square_value(2))
print(square_value(-4))
```

```python
def is_prime(n):
    if n in [2, 3]:  #si n es 2 o 3, es primo.
        return True
    if (n == 1) or (n % 2 == 0):  #si n es 1 o múltiplo de 2, no es primo.
        return False
    r = 3
    while r * r <= n:  #arrancando en 3, si r * r es mayor a n, se corta el loop.
        if n % r == 0:  #si es divisible por r, no es primo.
            return False
        r += 2 #si hasta acá dio true, entra al siguiente loop sumando 2
    return True
print(is_prime(49), is_prime(79))
```

```python
def persona(nombre, apellido):
    print(nombre, apellido)


# Keyword arguments
persona(nombre='Miguel', apellido='Cervantes')
```

```python
def myFun(*words):  #si no sabés cuántos argumentos va a tener tu función, se agrega el * adelante.
    """This function takes variable number of arguments"""
    print("Output is:")
    for word in words:
        print(word)

myFun('Hola', 'Soy', 'Frodo')
```

```python
#Keyword arguments
#Si no sabés cuántos keyward arguments va a tener la función, se agrega **
def myFun(**kwargs):
    for key, value in kwargs.items():
        print("%s == %s" % (key, value))


# Driver code
myFun(first='Breaking Bad', mid='GOT', last='Friends')
```

```python
#recursion in function: cuando una función llama a sí misma:
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(4))
```

```python
def swap(x, y):
    temp = x
    x = y
    y = temp

# Driver code
x = 2
y = 3
swap(x, y)
print(x)
print(y)
```

```python
def lucky_number(nums):
    for num in nums:
        if num % 7 == 0:
            return True
    # We've exhausted the list without finding a lucky number
    return False

a = (15, 24, 31, 35)
print(lucky_number(a))
```
